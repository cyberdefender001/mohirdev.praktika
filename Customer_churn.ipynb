```python
data = df[['Churn', 'Tenure', 'Complain', 'DaySinceLastOrder', 'CashbackAmount', 'MaritalStatus', 'Gender']]
```

```python
missing_amounts = data.isnull().sum().sum()
missing_rows = data[data.isnull().any(axis=1)].shape[0]
print(f"Mavjud bo'lmagan qiymatlar: {missing_amounts}")
print(f"Mavjud bo'lmagan qatorlar: {missing_rows}")
print(f"Mavjud bo'lmagan qatorlar foizda: {np.round(missing_rows/len(data)*100)}")
```

```python
data = data.dropna()
data.shape
```

```python
data['Churn'].value_counts()/len(data)*100
```

ML ga tayyorgarlik

```python
encoded = pd.get_dummies(data)
encoded.head()
```

```python
X = encoded.drop('Churn', axis=1)
y = encoded['Churn']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

MachineLearning

```python
#modelni yaratish
LR_model = LogisticRegression()
LR_model.fit(X_train, y_train)

#modelni baholash
y_pred = LR_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model aniqligi: ", metrics.accuracy_score(y_test, y_pred))

#confusion_matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='g')
plt.show()
```

```python

fpr, tpr, theresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

Support Vector Machine

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)
print("Classification report", metrics.classification_report(y_test, y_pred))
print('Accuracy: ', metrics.accuracy_score(y_test, y_pred))
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='g')
plt.show()
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

Decision Tree

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)
print("class_report", metrics.classification_report(y_test, y_pred))
print(metrics.accuracy_score(y_test, y_pred))
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='g')
plt.show()
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

```python
cols = encoded.drop('Churn', axis=1).columns

plt.figure(figsize=(30,20))
plot_tree(tree_model, feature_names=cols, filled=True)
plt.show()
```
